# Customer Churn Analysis — Demo Notebook

This notebook is a guided walkthrough of the churn analysis pipeline implemented in `src/`. It demonstrates cleaning, EDA, segmentation, CLV estimation, and visualization step by step.

Run this notebook after installing dependencies with `pip install -r requirements.txt`.

```python
from pathlib import Path

from src import clv, eda, segmentation, visualization
from src.data_cleaning import clean_pipeline

RAW_PATH = Path("../data/raw/sample_churn_data.csv")
```

## 1. Data Cleaning

Load the raw CSV and run the full cleaning pipeline.

```python
df, quality_report = clean_pipeline(RAW_PATH)
print(quality_report)
df.head()
```

## 2. Churn and Retention KPIs

Compute the overall churn rate and retention rate.

```python
print("Churn rate (%):", eda.churn_rate(df))
print("Retention rate (%):", eda.retention_rate(df))
```

## 3. Churn by Contract Type, Tenure, and Charges

```python
churn_by_contract = eda.churn_by_column(df, "contract_type")
churn_by_contract
```

```python
tenure_summary = eda.tenure_summary(df)
monthly_charges_summary = eda.monthly_charges_summary(df)
tenure_summary
```

## 4. Customer Segmentation

Assign a simple, rule-based risk segment to each customer.

```python
df = segmentation.add_risk_segment(df)
segment_summary = segmentation.segment_summary(df)
segment_summary
```

## 5. Customer Lifetime Value (CLV)

```python
df = clv.add_estimated_clv(df)
clv_by_segment = clv.clv_summary_by_segment(df)
clv_by_segment
```

## 6. Visualizations

Generate and save the standard chart set to the `images/` folder.

```python
IMAGES_DIR = Path("../images")

visualization.plot_churn_rate_overall(df, IMAGES_DIR / "churn_rate_overall.png")
visualization.plot_churn_by_contract_type(churn_by_contract, IMAGES_DIR / "churn_by_contract_type.png")
visualization.plot_churn_by_segment(segment_summary, IMAGES_DIR / "churn_by_segment.png")
```
